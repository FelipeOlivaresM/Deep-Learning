```python
#Importar bibliotecas
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sb
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
from sklearn.preprocessing import StandardScaler
from scipy import stats
%matplotlib inline

import warnings
warnings.filterwarnings('ignore')
```

```python
#Cargar data de kaggle Fifa19

df = pd.read_csv("data.csv")
```

```python
#Eliminar primera columna sin información relevante
df.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
"""Función para convertir en numeros los salarios y valores cada jugador"""
def convertValue(x):
    value = str(x).replace('€', '')
    if('M' in str(value)):
        value = str(value).replace('M', '')
        value = float(value) * 1000000
    elif('K' in str(value)):
        value = str(value).replace('K', '')
        value = float(value) * 1000
    return float(value)

df['Value'] = df['Value'].apply(convertValue)
df['Wage'] = df['Wage'].apply(convertValue)
```

```python
"""Convertir las posiciones a varibles numericas"""

dataf_Fifa = pd.get_dummies(dataf_Fifa, columns=['Position'], drop_first=True)
dataf_Fifa.head()
```

```python
dataf_Fifa = dataf_Fifa.fillna(method='bfill', limit=100)
```

```python
#print(dataf_Fifa.isnull().sum())
```

```python
""" Separación entre labels y features para el entrenamiento del modelo """

labels = dataf_Fifa['Value']
del dataf_Fifa['Value']
del dataf_Fifa['Name']
```

```python
#features = dataf_Fifa.drop(['Name', 'Value'], axis=1)
```

```python
""" Se utiliza la herramienta de Scikit Learn para dividir la date en entrenamiento y prueba"""

Xtrain, Xtest, ytrain, ytest = train_test_split(dataf_Fifa, labels, test_size=0.3, random_state=101)
```

```python
""" Se instancia el metodo de la regresión lineal """
ScalLinearRegr = StandardScaler()
Xtrain = ScalLinearRegr.fit_transform(Xtrain)
Xtest = ScalLinearRegr.fit_transform(Xtest)
```

```python
def predict_Value(Xtrain, Xtest, ytrain):
    lm = LinearRegression()
    lm.fit(Xtrain, ytrain)
    values = lm.predict(Xtest)
    return values
```

```python
predicted_wages1 = predict_Value(Xtrain, Xtest, ytrain)
```

```python
print('Mean Absolute Error : ' + str(metrics.mean_absolute_error(ytest, predicted_wages1)))
print('Mean Squared Error : ' + str(metrics.mean_squared_error(ytest, predicted_wages1)))
print('Root Mean Squared Error : ' + str(np.sqrt(metrics.mean_squared_error(ytest, predicted_wages1))))
```

```python
sb.distplot(ytest-predicted_wages1, bins=200, hist_kws={'color':'r'}, kde_kws={'color':'y'})
plt.xlim(-10000000, 10000000)
```

### Delimitación de los jugadores subvalorados para compra inmediata y posterior venta en 2 años

```python
""" Se opta por realizar una organización en el dataframe enfocandose
en las principales 3 variables que implican al precio del jugador.

Es importante resaltar que se opta por jugadores que sean jovenes y a su vez cuenten con un gran potencial,
lo cual represente la inversión a largo plazo"""

dataf_Fifa = df[['Name', 'Age', 'Overall', 'Potential', 'Value', 'Wage', 'Reactions', 'Special', 'Composure', 'International Reputation', 'Position']]
dataf_Fifa.head()
df_byAPV = dataf_Fifa.sort_values(['Age','Potential','Value'],ascending=[True, False, False,])
df_byAPV.head()
```

```python
""" Con el fin de obtener una ganancia e inversión no tan riesgosa para el equipo
se opta por filtrar jugadores que posean un potencial superior a la metrica de 78
y posteriormente su valor monetario no sea superior a los 200000 euros"""

def selectPlayers(potential,value,age):
    
    df_byPot = df_byAPV[(df_byAPV['Potential'] >= potential) & (df_byAPV['Value'] < value) 
                        & (df_byAPV['Age'] < age)]
    print("Cantidad de jugadores recolectados: {0}".format(len(df_byPot.index)))
    return df_byPot


def selectPlayersforPos(dataframe,position):
    
    df_byPosition = dataframe[dataframe.Position.isin([position])]
    print("Cantidad de jugadores recolectados: {0}".format(len(df_byPosition.index)))
    
    
    return df_byPosition
```

```python
"""Mostrar todos los jugadores para fichar"""

""" Se obtiene un total de 25 jugadores basado en estos parametros
si se desea de acuerdo al presupuesto del equipo e interes en posiciones
se puede obtener una mayor cantidad de jugadores, hay que tener en cuenta
que los jugadores que no posean un valor hacen referencia a jugadores que actualmente 
no se encuentran fichados por ningún equipo"""


df_byPot = selectPlayers(78,200000,26)
df_byPot.head()
```

```python
"""Mostrar todos los jugadores para fichar por posición requerida"""

df_byPos = selectPlayersforPos(df_byPot,'GK')
df_byPos.head()
```
